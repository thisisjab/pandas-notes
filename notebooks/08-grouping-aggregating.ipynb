What if we want to see results based on country? Like what social media is popular in which country. We need grouping data. Grouping data consits of three steps: splitting, applying functions, and combining results.

### Grouping

```python
country_group = survey_df.groupby(['Country'])
```

```python
country_group.get_group('Iran')
```

```python
country_group['SocialMedia'].value_counts().head(20)
```

```python
country_group['SocialMedia'].value_counts(normalize=True).head(20).loc['Afghanistan']
```

```python
country_group['ConvertedComp'].median().loc[['Afghanistan', 'Iran', 'Germany']]
```

```python
country_group['ConvertedComp'].agg(['mean', 'median'])
```

```python
filt = survey_df['Country'] == 'India'
```

```python
survey_df[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

Now we want to do the same thing on groups.

```python
country_group['LanguageWorkedWith'].apply(lambda country: country.str.contains('Python').sum())
```
